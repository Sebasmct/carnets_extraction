```python
import cv2
import pytesseract
import matplotlib.pyplot as plt
```

```python
img_file = "data/carnet.jpeg"
img_h = cv2.imread(img_file)
plt.imshow(img_h)
```

```python
img_name = img_h[500:600,:]
img_role = img_h[600:640,:]
img_id = img_h[640:680,:300]

fig, ax = plt.subplots(3,1)

ax[0].imshow(img_name)
ax[1].imshow(img_role)
ax[2].imshow(img_id)
```

For better results use psm:
- 6: Assume a Single Uniform Block of Text
- 11: Sparse Text: Find as Much Text as Possible in No Particular Order

Take into account:
- 7: Treat the Image as a Single Text Line
- 10: Treat the Image as a Single Character

```python
config = "--psm 11 --oem 1"
name = pytesseract.image_to_string(img_name, config=config+""" -c tessedit_char_blacklist=`~!@#$%^&*()-_=+[]{}|\\;:\'",.<>?/`
""")
role = pytesseract.image_to_string(img_role, config=config+""" -c tessedit_char_blacklist=`~!@#$%^&*()-_=+[]{}|\\;:'",.<>?/`
""")
id_number = pytesseract.image_to_string(img_id, config=config+" -c tessedit_char_whitelist=0123456789")

print(name)
print(role)
print(id_number)
```

```python
name2 = [x for x in name.split("\n") if len(x)>2]
role2 = [x for x in role.split("\n") if len(x)>2]

print(name2)
print(role2)
print(id_number)
```
